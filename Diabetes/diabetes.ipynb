```python
import seaborn as sb
import matplotlib as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,normalize
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import normalize
```

```python
#reading data as pandas 
data=pd.read_csv("diabetes.csv")

#data.
#data.info()
#checking data 
#data.describe()
data.isnull().sum()
```

```python

data.mean()
```

```python
#spliting features and target
x=data.drop(columns="Outcome",axis=1)
y=data["Outcome"]
print(x)
```

```python
#whhat the algorithm will spot :glucose and insulin levels are important featrues  
data.groupby("Outcome").mean()
```

Data standalization


```python
#standalize the data (data processing) why?: Data standardization helps improve the quality of your data by transforming and standardizing it 
#helps reduce cost and imporve the algorithm s performance and accuracy (works on gaussian distribution)
```

Scaling


```python
scaler=StandardScaler()
scaler.fit(x)
standarlized_data=scaler.transform(x)

print(x)
```

```python
#split data into train and test data
x_train,x_test,y_train,y_test=train_test_split(standarlized_data,y,test_size=0.2)
print ("x_train",x_train.shape)
print ("x_test",x_test.shape)
print ("y_train shape",y_train.shape)
print ("y_test",y_test.shape)
```

```python
Model2=LogisticRegression(max_iter=1500)
Model2.fit(x_train,y_train)

```

Accuracy on Training data

```python
predicted=Model2.predict(x_train)
train_accuracy=accuracy_score(predicted,y_train)
print("accuracy score on training data ",train_accuracy)
```

Accuracy on test data


```python
predicted=Model2.predict(x_test)
test_accuracy=accuracy_score(predicted,y_test)
print("accuracy score on test data",test_accuracy)
```
