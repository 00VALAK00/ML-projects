```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
data=pd.read_csv("full_filled_stroke_data.csv")
```

```python
data.head()
```

```python
labelencoder=preprocessing.LabelEncoder()
data["gender"]=labelencoder.fit_transform(data["gender"])
data["gender"].unique()
print(data["gender"])
```

```python
data["smoking_status"]=labelencoder.fit_transform(data["smoking_status"])
data["smoking_status"].unique()
```

```python
data["ever_married"]=labelencoder.fit_transform(data["ever_married"])
data["ever_married"].unique()
```

standarlize the data

```python
data.groupby("stroke").mean()
```

```python
data["Residence_type"]=labelencoder.fit_transform(data["Residence_type"])
data["Residence_type"].unique()
```

```python
y=data["stroke"]
x=data.drop(columns=["stroke","work_type"],axis=1)
print (data)
```

```python
scaler=StandardScaler()
scaler.fit(x)
standarlized_data=scaler.fit_transform(x)
```

Spliting datasets


```python
x_train,x_test,y_train,y_test=train_test_split(standarlized_data,y,test_size=0.3 )
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
Model=LogisticRegression(max_iter=1500)
Model.fit(x_train,y_train)
```

```python
predicted=Model.predict(x_test)

test_accuracy=accuracy_score(predicted,y_test)
print("accuracy score on test data",test_accuracy)
```

Here we check the Accuracy of training data

```python
predicted=Model.predict(x_train)
train_accuracy=accuracy_score(predicted,y_train)
print("accuracy score on training data ",train_accuracy)
```
